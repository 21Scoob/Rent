```python
import pandas as pd 

df = pd.read_csv('rent.csv')

dp = df.dropna()

print(dp)

print(df.to_string())
```

From here we can see that we have no empty cells and everything is full

```python
keyword = "National currency"

df = df[~df['currency'].str.contains(keyword, case=False, na=False)]

print(df.to_string())
```

I observed that in some countries the currency value was set on National Currency and Euro, so for simplicity and no disruptancy i decided that i should delete all the rows with the currency status "National Currency"

```python
new_df = df.drop(['OBS_FLAG', 'CONF_STATUS'], axis=1)
print(new_df.to_string())
```

Deleted the last two columns with no values for more readability 

```python
df = df.rename(columns={''})
```

```python
df_mean_2023 = df[df["TIME_PERIOD"] == 2023].groupby("geo")["OBS_VALUE"].mean()

print(df_mean_2023)
```

```python
df_mean_2013 = df[df["TIME_PERIOD"] == 2013].groupby("geo")["OBS_VALUE"].mean()

print(df_mean_2013) 
```

```python
df_mean_2018 = df[df["TIME_PERIOD"] == 2018].groupby("geo")["OBS_VALUE"].mean()

print(df_mean_2018) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.kdeplot(
    data=new_df[new_df['region'] == 'Est'],
    x='OBS_VALUE',
    fill=True,
    color='royalblue',
    label='Est'
)
sns.kdeplot(
    data=new_df[new_df['region'] == 'Vest'],
    x='OBS_VALUE',
    fill=True,
    color='salmon',
    label='Vest'
)

plt.title('Distribuția densității chiriilor: Est vs Vest')
plt.xlabel('Chirie (OBS_VALUE)')
plt.ylabel('Densitate')
plt.legend()
plt.show()
```
